### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import numpy as np
from collections.abc import Iterable

# Sample Data
data = np.random.randn(1000)  # Generating random values for the histogram
bins = 20  # Number of bins

# Error Handling
if not isinstance(data, Iterable):
    raise TypeError("Data must be an iterable (list, tuple, or NumPy array).")
if not isinstance(bins, int) or bins <= 0:
    raise ValueError("The 'bins' parameter must be a positive integer.")

# Create Histogram
plt.hist(data, bins=bins, color='skyblue', edgecolor='black', alpha=0.7)

# Labels and Title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Customized Histogram')

# Show the plot
plt.show()
```
